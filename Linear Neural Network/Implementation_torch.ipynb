```python
import numpy as np 
import torch 
import torch.nn as nn 
```

### Model


```python
class LNN(nn.Module):
    
    
    def __init__(self,input,output):   
        super(LNN, self).__init__()
        self.linear = nn.Linear(input, output)
        
    def forward(self, x):
        out = self.linear(x)
        return out
    
    
    # backward is handled automatically by PyTorch's autograd
                        
        
```

### DATA

```python
np.random.seed(0)
X = np.random.randn(100, 2)
true_W = np.array([[3],[5]])
true_b = 10
Y = X @ true_W + true_b + 0.2*np.random.randn(100, 1)   # add slight noise

X = torch.tensor(X, dtype=torch.float32)
Y = torch.tensor(Y, dtype=torch.float32)
```

```python
model = LNN(2,1)  # 2 inputs, 1 outputs


epochs = 2000
lr = 0.01

model = LNN(2,1)


print(model.linear.weight.data)
print(model.linear.bias.data)


criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=lr)


for epoch in range(epochs):

    
    out = model.forward(X)
    
    loss = criterion(out,Y)
    
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    
    if (epoch+1) % 200 == 0:
        print(f"Epoch {epoch+1}/{epochs}  Loss = {loss:.4f}")
        
print("\nLearned W:\n", model.linear.weight.data)
print("Learned b:\n", model.linear.bias.data)
```
